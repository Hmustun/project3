```python
# Importing modules
from yahoofinancials import YahooFinancials
import pandas as pd
```

```python
# Selecting Apple Inc. ticker
ticker = 'AAPL'
yf = YahooFinancials(ticker)
```

## Income Statement

```python
# Extracting Income Statement data
income_statement = yf.get_financial_stmts('annual', 'income')

aapl_is = income_statement['incomeStatementHistory']['AAPL']
aapl_is
```

```python
# The data is in the form of a dictionary and it needs to be converted into a DataFrame

df_list = []

for d in aapl_is:
    df_list.append(pd.DataFrame.from_dict(d, orient = 'index'))

df_is = pd.concat(df_list)
df_is
```

# Balance Sheet

```python
# Extracting Balance Sheet data
balance_sheet = yf.get_financial_stmts('annual', 'balance')

aapl_bs = balance_sheet['balanceSheetHistory']['AAPL']
aapl_bs
```

```python
# The data is in the form of a dictionary and it needs to be converted into a DataFrame

df_list = []

for d in aapl_bs:
    df_list.append(pd.DataFrame.from_dict(d, orient = 'index'))

df_bs = pd.concat(df_list)
df_bs
```

# Exporting data to Excel

```python
with pd.ExcelWriter('AAPL Financial Statemetns.xlsx') as writer:

    df_is.to_excel(writer, sheet_name = 'Income Statement', index = True)
    df_bs.to_excel(writer, sheet_name = 'Balance Sheet', index = True)
```
